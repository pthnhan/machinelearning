```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
```

# **Generate data**

```python
# height (cm)
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
# weight (kg)
y = np.array([[ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T
```

```python
# Visualize data 
plt.plot(X, y, 'ro')
plt.axis([140, 190, 45, 75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

# **Manual**

```python
# Building Xbar 
one = np.ones((X.shape[0], 1))
Xbar = np.concatenate((one, X), axis = 1)
# Calculating weights of the fitting line 
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)
w = np.dot(np.linalg.pinv(A), b)
```

# **Sklearn**

```python
from sklearn import linear_model

# fit the model by Linear Regression
regr = linear_model.LinearRegression(fit_intercept=True) # fit_intercept = False for calculating the bias
regr.fit(Xbar, y)
```

```python
regr.coef_
```

# **Compare two results**

```python
regr.intercept_
```

```python
print( 'Solution found by scikit-learn  : ', regr.coef_ )
print( 'Solution found by manual: ', w.T)
```

# **Visualize fitted line**

```python
# Generate the line
x_line = np.linspace(140, 190, 100)
y_line = w[0] + w[1] * x_line

# Reshape X to match the shape of x_line
X_reshaped = X.reshape(-1)

# Plot the data points and the line
plt.plot(X_reshaped, y.reshape(-1), 'ro', label='Data')
plt.plot(x_line, y_line, label='Line')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.show()
```
